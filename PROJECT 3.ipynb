# BIVARIATE ANALYSIS

```python
# numerical vs numerical
#GrLivArea vs SalePrice
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title("GrLivArea vs SalePrice")
plt.show()
```

There is a strong positive relationship between GrLivArea and SalePrice. As the living area increases, the sale price of houses also increases, indicating that larger houses are generally more expensive.

```python
# numercial vs numerical
# TotalBsmtSF vs SalePrice
sns.scatterplot(x='TotalBsmtSF', y='SalePrice', data=df)
plt.title("TotalBsmtSF vs SalePrice")
plt.show()
```

Houses with larger basement areas tend to have higher sale prices, although the relationship is weaker compared to GrLivArea.

```python
# Numerical vs Categorical
# OverallQual vs SalePrice
sns.boxplot(x='OverallQual', y='SalePrice', data=df)
plt.title("OverallQual vs SalePrice")
plt.show()

```

SalePrice increases significantly with higher OverallQual ratings, indicating that overall construction quality has a strong impact on house prices.

```python
#Categorical vs Categorical
# Neighborhood vs HouseStyle
sns.countplot(x='Neighborhood', hue='HouseStyle', data=df)
plt.xticks(rotation=90)
plt.title("Neighborhood vs HouseStyle")
plt.show()
```

Certain house styles are more prevalent in specific neighborhoods, suggesting that location influences housing structure patterns.

```python
# Year-based Bivariate Analysis
# YearBuilt vs SalePrice
sns.scatterplot(x='YearBuilt', y='SalePrice', data=df)
plt.title("YearBuilt vs SalePrice")
plt.show()

```

Newer houses generally command higher sale prices, showing a positive trend between construction year and price.

```python
# Correlation with Target
df.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)

```

obsrevation:- Features such as OverallQual, GrLivArea, and GarageArea show strong positive correlation with SalePrice.

```python
# Correlation Heatmap (Multiple Numerical Variables)
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

The heatmap shows that OverallQual, GrLivArea, and GarageArea have strong positive correlations with SalePrice, indicating these are key factors influencing house prices.

```python
# Living Area vs Price (colored by Quality)
sns.scatterplot(
    x='GrLivArea',
    y='SalePrice',
    hue='OverallQual',
    data=df
)
plt.title("GrLivArea vs SalePrice by OverallQual")
plt.show()
```

For similar living areas, houses with higher overall quality tend to have higher sale prices, highlighting the combined effect of size and quality.

```python
# SalePrice across Neighborhood and HouseStyle
sns.boxplot(
    x='Neighborhood',
    y='SalePrice',
    hue='HouseStyle',
    data=df
)
plt.xticks(rotation=90)
plt.title("SalePrice by Neighborhood and HouseStyle")
plt.show()
```

observation- SalePrice varies significantly across neighborhoods, and this variation is further influenced by house style, indicating the combined impact of location and design.

```python
# Pairplot (Key Features)
important_cols = ['SalePrice','GrLivArea','OverallQual','GarageArea']
sns.pairplot(df[important_cols])
plt.show()

```

Pairwise plots reveal strong linear relationships between SalePrice and features like GrLivArea and OverallQual.

```python
# YearBuilt & Quality Combined Effect
sns.scatterplot(
    x='YearBuilt',
    y='SalePrice',
    hue='OverallQual',
    data=df
)
plt.title("YearBuilt vs SalePrice by OverallQual")
plt.show()

```

Newer houses with higher quality ratings tend to achieve the highest sale prices, showing a combined influence of age and quality.
